## Lab | Random variable distributions

```python
#import libraries
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

* Get the numerical variables from our dataset.

```python
customer_df = pd.read_csv('files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
```

```python
#rename columns
customer_df.columns = [i.lower().replace(' ', '_') for i in customer_df.columns]
customer_df = customer_df.rename(columns = {'employmentstatus' : 'employment_status'})
customer_df['effective_to_date'] = pd.to_datetime(customer_df['effective_to_date'])
#set customer as an index
customer_df.set_index('customer', inplace = True)
customer_df.head()
```

```python
#select numerical columns
numerical = customer_df.select_dtypes(np.number)
numerical.columns
```

* Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.

```python
fig, ax = plt.subplots(2,4, figsize = (16,8))
sns.distplot(numerical['customer_lifetime_value'], kde = False, ax = ax[0, 0]) 
sns.distplot(numerical['income'], ax = ax[0, 1]) 
sns.distplot(numerical['monthly_premium_auto'], ax = ax[0, 2]) 
sns.distplot(numerical['months_since_last_claim'], ax = ax[0, 3]) 
sns.distplot(numerical['months_since_policy_inception'], ax = ax[1, 0]) 
sns.distplot(numerical['number_of_open_complaints'], ax = ax[1, 1]) 
sns.distplot(numerical['number_of_policies'], ax = ax[1, 2]) 
sns.distplot(numerical['total_claim_amount'], ax = ax[1, 3])
plt.tight_layout()
#it seems that some of them are exponential (e.g number of complaints and policies)
#but we will see more clear after transformations
```

* Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.

```python
#apply boxcox and check results
boxcox_transform = numerical.copy()
for col in boxcox_transform:
    boxcox_transform[col] = np.where(boxcox_transform[col] <= 0, np.NAN, boxcox_transform[col])
    boxcox_transform[col] = boxcox_transform[col].fillna(boxcox_transform[col].mean())
    fitted_data, fitted_lambda = stats.boxcox(boxcox_transform[col])
    boxcox_transform[col] = fitted_data
```

```python
fig, ax = plt.subplots(2,4, figsize = (16,8))
sns.distplot(boxcox_transform['customer_lifetime_value'], ax = ax[0, 0]) 
sns.distplot(boxcox_transform['income'], ax = ax[0, 1]) 
sns.distplot(boxcox_transform['monthly_premium_auto'], ax = ax[0, 2]) 
sns.distplot(boxcox_transform['months_since_last_claim'], ax = ax[0, 3]) 
sns.distplot(boxcox_transform['months_since_policy_inception'], ax = ax[1, 0]) 
sns.distplot(boxcox_transform['number_of_open_complaints'], ax = ax[1, 1]) 
sns.distplot(boxcox_transform['number_of_policies'], ax = ax[1, 2]) 
sns.distplot(boxcox_transform['total_claim_amount'], ax = ax[1, 3])
plt.tight_layout()
```

```python
#apply logarithmic transformation and check results
log_transform = numerical.copy()
transform = lambda x: np.log(x) if np.isfinite(x) and x!=0 else np.NAN
for col in log_transform:
    log_transform[col] = log_transform[col].apply(transform).fillna(log_transform[col].mean())
```

```python
fig, ax = plt.subplots(2,4, figsize = (16,8))
sns.distplot(log_transform['customer_lifetime_value'], ax = ax[0, 0]) 
sns.distplot(log_transform['income'], ax = ax[0, 1]) 
sns.distplot(log_transform['monthly_premium_auto'], ax = ax[0, 2]) 
sns.distplot(log_transform['months_since_last_claim'], ax = ax[0, 3]) 
sns.distplot(log_transform['months_since_policy_inception'], ax = ax[1, 0]) 
sns.distplot(log_transform['number_of_open_complaints'], ax = ax[1, 1]) 
sns.distplot(log_transform['number_of_policies'], ax = ax[1, 2]) 
sns.distplot(log_transform['total_claim_amount'], ax = ax[1, 3])
plt.tight_layout()
```

__Outcome:__

__customer_lifetime_value__ seems to have exponential distribution. However after boxcox and log transformation (they show almost same result) the distribution becomes more normal.

__income / monthly_premium_auto / months_since_last_claim__ no meaning to apply any trasformation. It can be considered as neither normal nor exponential..

__months_since_policy_inception__ is more normal with log transformation.

__number_of_open_complaints / number_of_policies__ the transformation didn't change them to normal. In origin they can be considered as exponential.

__total_claim_amount__ it seems that the transformations were more useful in this case. Both boxcox and logarithmic made the distribution normal.
